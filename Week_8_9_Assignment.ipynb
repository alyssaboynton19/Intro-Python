```
import random
import urllib.request

def choose_word():
  file_name = "https://raw.githubusercontent.com/mlepinski/Python-Worksheets/master/coolWords.txt"
  my_file = urllib.request.urlopen(file_name)
  long_string = my_file.read()
  long_string = long_string.decode("utf-8")
  word_list = long_string.split("\n")
  return random.choice(word_list)
```

```
def print_body(count):
  if count == 1:
    print("   (_)   ")
  if count == 2:
    print("   (_)   ")
    print("    |    ")
    print("    |    ")
  if count == 3:
    print("   (_)   ")
    print("   /|    ")
    print("  / |    ")
  if count == 4:
    print("   (_)   ")
    print("   /|\   ")
    print("  / | \  ")
  if count == 5:
    print("   (_)   ")
    print("   /|\   ")
    print("  / | \  ")
    print("   /     ")
    print("  /      ")
  if count == 6:
    print("   (_)   ")
    print("   /|\   ")
    print("  / | \  ")
    print("   / \   ")
    print("  /   \  ")   
  if count == 7:
    print("   (_)   ")
    print("   /|\   ")
    print("  / | \  ")
    print("   / \   ")
    print("  /   \  ") 
    print("--     --")
```

```
hangman_word = choose_word().lower()


hangman_list = []              
for letter in hangman_word:
  hangman_list.append('_ ')

count = 0
wrong_guesses = []
guesses_list = []
hangman_string = ''

print('''Welcome to my hangman game!
Fair warning, all the words in this game are the names of Pokemon...
So I hope you're ready to catch 'em all!''')

while True:
  print(hangman_string)
  guessed_letter = input('Please enter a letter: ').lower()

  while len(guessed_letter) > 1 or guessed_letter in '0123456789' or guessed_letter in guesses_list: 
    if len(guessed_letter) > 1:
      print('That isn\'t a single letter.')
      guessed_letter = input('Please enter a letter: ').lower()
    if guessed_letter in '0123456789':
      print('That isn\'t a letter.')
      guessed_letter = input('Please enter a letter: ').lower()
    if guessed_letter in guesses_list:
      print('You already guessed that!')
      guessed_letter = input('Please enter a letter: ').lower()

  guesses_list.append(guessed_letter)
  
  if not guessed_letter in hangman_word:
    count = count + 1
    if count == 7:
      print_body(count)
      print('Oh no, you lost! The word was', hangman_word)
      break
    wrong_guesses.append(guessed_letter)
    print('Oh no, you guessed an incorrect letter!')
    print_body(count)
    lives = 7 - count
    print('You have ' + str(lives) + ' lives left.') 
    print('Here are your incorrect guesses so far:', wrong_guesses)
  else:
    check_list = list(hangman_word)
    for letter in range(len(hangman_word)):
      if not guessed_letter in check_list:
        break
      num = check_list.index(guessed_letter)
      hangman_list[num] = guessed_letter
      check_list[num] = '-'
      hangman_string = ''
      hangman_string = hangman_string.join(hangman_list)
    if hangman_string == hangman_word:
      print('You won the game! You correctly guessed', hangman_word.capitalize())
      break
    print('You guessed correctly!')
```
